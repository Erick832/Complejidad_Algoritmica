```python
import pandas as pd
import numpy as np
import networkx as nx
import math
from heapq import heappush as push, heappop as pop
import json
```

```python
def haversine(cp1, cp2):
  la1, lo1 = float(cp1['LATITUD']), float(cp1['LONGITUD'])
  la2, lo2 = float(cp2['LATITUD']), float(cp2['LONGITUD'])
  
  lo1, la1, lo2, la2 = map(math.radians, [lo1, la1, lo2, la2])
  dlo = lo2 - lo1
  dla = la2 - la1
  a = math.sin(dla/2)**2 + math.cos(la1) * math.cos(la2) * math.sin(dlo/2)**2
  c = 2 * math.asin(math.sqrt(a))
  r = 6371

  return round(c * r, 2)
def prim(G, s):
  for u in G:
    G.nodes[u]['visited'] = False
    G.nodes[u]['cost'] = math.inf
    G.nodes[u]['path'] = -1
  
  q = [(0, s)]
  while q:
    _, u = pop(q)
    if not G.nodes[u]['visited']:
      G.nodes[u]['visited'] = True
      for v in G.neighbors(u):
        if not G.nodes[v]['visited']:
          w = G.edges[u, v]['weight']
          if w < G.nodes[v]['cost']:
            G.nodes[v]['cost'] = w
            G.nodes[v]['path'] = u
            push(q, (w, v))
  
  path = [-1]*G.number_of_nodes()
  lista=[]
  for i,j in enumerate (G.nodes):
    path[i] = G.nodes[j]['path']
  return path

def create_Graph(distrito,columna):
  Grap = nx.Graph()
  col = columna
  for i, cp1 in distrito.iterrows():
    Grap.add_node(i, label=cp1[col])

  for i, cp1 in distrito.iterrows():
    for j, cp2 in distrito.iterrows():
      if cp1[col] != cp2[col]:
        Grap.add_edge(i, j, weight=haversine(cp1, cp2))
  return Grap

def mostrar_Lista(G,ph,sta):
  lista_nodos=G.nodes
  lista_nodos=list(lista_nodos)   
  #start=lista_nodos[s]
  lista=[]
  start=sta
  lista.append(start)
  while (len(lista)<G.number_of_nodes()):
    i=0
    for i,j in enumerate(ph):
      if (start==j):
        break;
    lista.append(lista_nodos[i])
    start=lista_nodos[i]
  return lista
def cambiar_a_nombre(cam,G):
  lista_nombres=[]
  for i in cam:
    lista_nombres.append(G.nodes[i]['label'])
  return lista_nombres

def cambiarformato(camino,poblacionesDF):
  ResponsePath=list()
  for i in camino:
    poblacionesDF.iloc[i]['CENTRO POBLADO']
    ResponsePath.append({"cp":poblacionesDF.iloc[i]['CENTRO POBLADO'],"lat":poblacionesDF.iloc[i]['LATITUD'],"lon":poblacionesDF.iloc[i]['LONGITUD']})
  return ResponsePath

def Get_lista_provincias():
  poblacionesDF = pd.read_csv(url)
  pdDepartamentos=poblacionesDF['DEPARTAMENTO'].unique()
  departamentos=[]
  G_d=nx.Graph()
  for i in pdDepartamentos:
    pf=poblacionesDF[poblacionesDF['DEPARTAMENTO']==i]
    departamentos.append(pf.head(1))
  lis=[]
  for i in departamentos:
    lis.append(*i.index)# GENERO UNA LISTA
  departamentos=poblacionesDF.iloc[lis]#SELECIONO UNA TABLA SOLO CON LOS DATOS DE LA LISTA
  G_d=create_Graph(departamentos,'DEPARTAMENTO')
  lista_nodos=G_d.nodes()
  lista_nodos=list(lista_nodos)
  start=lista_nodos[22]#->tumbes
  path = prim(G_d,start)
  camino=mostrar_Lista(G_d,path,start)
  camino_nombres=cambiar_a_nombre(camino,G_d)

  departamentos_resultantes=[]
  for element in camino_nombres :
      if element not in departamentos_resultantes:
          departamentos_resultantes.append(element)
  departamentos_resultantes#->LISTA CON LOS DEPARTAMENTOS
  #departamentos_resultantes.extend(['LORETO','PUNO','TACNA','MOQUEGUA','AREQUIPA','ICA','LIMA'])
  nomprovincias = poblacionesDF['PROVINCIA'].unique()
  provincias_resultantes=[]
  for departamen in departamentos_resultantes:
    provincias=dict()
    for nom in nomprovincias:
        pf=poblacionesDF[(poblacionesDF['DEPARTAMENTO']==departamen) & (poblacionesDF['PROVINCIA']==nom)&(poblacionesDF['NMI1']>0)]
        if not pf.empty:
          provincias_resultantes.append(pf.head(1)) 
  indices_de_provincias_por_orden=[]
  for p in provincias_resultantes:
      indices_de_provincias_por_orden.append(*p.index)

  secuencias_provincias=poblacionesDF.iloc[indices_de_provincias_por_orden]
  sc=secuencias_provincias['PROVINCIA'].unique()
  return sc
```

```python
url  = "https://raw.githubusercontent.com/lmcanavals/algorithmic_complexity/main/data/poblaciones.csv"
```

```python
def peru1():
  poblacionesDF = pd.read_csv(url)
  #nomprovincias = poblacionesDF['PROVINCIA'].unique()
  nomprovincias=Get_lista_provincias()
  supercamino=[]

  for nom_provincia in nomprovincias:
    provincias = dict()
    nomdistritos=list()
    distritos = dict()
    for nom in nomprovincias:
      provincias[nom] = poblacionesDF[poblacionesDF['PROVINCIA'] == nom]
    nomdistritos = provincias[nom_provincia]['DISTRITO'].unique()
    
    provincia = provincias[nom_provincia]
    for nom in nomdistritos:
      distritos[nom] = provincia[provincia['DISTRITO'] == nom]
    
    for Dist in nomdistritos:
      distrito = distritos[Dist]
      Grap=nx.Graph()
      Grap=create_Graph(distrito,'CENTRO POBLADO')
      lista_nodos=Grap.nodes()
      lista_nodos=list(lista_nodos)
      start=lista_nodos[0]
      path = prim(Grap,start)
      camino=mostrar_Lista(Grap,path,start)
      
      camino_nombres=cambiar_a_nombre(camino,Grap)
      suma=[]
      for i in Grap:
        if Grap.nodes[i]['cost']!=math.inf:
          suma.append(Grap.nodes[i]['cost'])
      temporal=cambiarformato(camino,poblacionesDF)
      if temporal:
        supercamino+=temporal
  return json.dumps(supercamino)
```
