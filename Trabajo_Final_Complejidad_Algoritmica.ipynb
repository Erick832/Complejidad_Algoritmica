```python
from algorithmic_complexity.aclib import graphstuff as gs
import pandas as pd
import numpy as np
import networkx as nx
import math
from heapq import heappush as push, heappop as pop
```

```python
def haversine(cp1, cp2):
  la1, lo1 = float(cp1['LATITUD']), float(cp1['LONGITUD'])
  la2, lo2 = float(cp2['LATITUD']), float(cp2['LONGITUD'])
  
  lo1, la1, lo2, la2 = map(math.radians, [lo1, la1, lo2, la2])
  dlo = lo2 - lo1
  dla = la2 - la1
  a = math.sin(dla/2)**2 + math.cos(la1) * math.cos(la2) * math.sin(dlo/2)**2
  c = 2 * math.asin(math.sqrt(a))
  r = 6371

  return round(c * r, 2)

def prim(G, s):
  for u in G:
    G.nodes[u]['visited'] = False
    G.nodes[u]['cost'] = math.inf
    G.nodes[u]['path'] = -1
  
  q = [(0, s)]
  while q:
    _, u = pop(q)
    if not G.nodes[u]['visited']:
      G.nodes[u]['visited'] = True
      for v in G.neighbors(u):
        if not G.nodes[v]['visited']:
          w = G.edges[u, v]['weight']
          if w < G.nodes[v]['cost']:
            G.nodes[v]['cost'] = w
            G.nodes[v]['path'] = u
            push(q, (w, v))
            #print(q)

  path = [-1]*G.number_of_nodes()
  lista=[]
  #for u in range(G.number_of_nodes()):
   # print(G.node[u])
  for i,j in enumerate (G.nodes):
    #print(i,"->",j)
    path[i] = G.nodes[j]['path']
  return path
def create_Graph(distrito):
  Grap = nx.Graph()
  #Grap2 = nx.Graph()
  col = 'CENTRO POBLADO'
  for i, cp1 in distrito.iterrows():
    Grap.add_node(i, label=cp1[col])
    #Grap2.add_node(i, label=cp1[col])

  for i, cp1 in distrito.iterrows():
    for j, cp2 in distrito.iterrows():
      if cp1[col] != cp2[col]:
        Grap.add_edge(i, j, weight=haversine(cp1, cp2))
        #Grap2.add_edge(i, j, weight=haversine(cp1, cp2))
  return Grap
def mostrar_Lista(G,ph):
  lista_nodos=G.nodes
  lista_nodos=list(lista_nodos)
  #print(lista_nodos)   
  start=lista_nodos[0]
  lista=[]
  lista.append(start)
  start=lista_nodos[0]
  while (len(lista)<G.number_of_nodes()):
    i=0
    for i,j in enumerate(ph):
      if (start==j):
        break;
    lista.append(lista_nodos[i])
    start=lista_nodos[i]
  return lista
def cambiar_a_nombre(cam,G):
  lista_nombres=[]
  #costo=0
  for i in camino:
    lista_nombres.append(G.nodes[i]['label'])
    #costo=costo+G.nodes[i]['cost']
  #print(costo)
  return lista_nombres

#Pasar al formato 
def cambiarformato(camino):
  ResponsePath=list()
  for i in camino:
    #print(i)
    poblacionesDF.iloc[i]['CENTRO POBLADO']
    ResponsePath.append({"cp":poblacionesDF.iloc[i]['CENTRO POBLADO'],"lat":poblacionesDF.iloc[i]['LATITUD'],"lon":poblacionesDF.iloc[i]['LONGITUD']})
  print(ResponsePath)
```
